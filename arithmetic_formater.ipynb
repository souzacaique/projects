```python
problems = ["32 + 6980", "3801 - 2", "45 + 43", "123 + 49"]

def arithmetic_arranger(problems, result = False):    
    #------------------------------------------------------------------dicionario que tem como chave o simbolo de operação e como valor o resultado de lambda
    ops = {"+": (lambda x,y: x+y), "-": (lambda x,y: x-y)}
    import re
    #------------------------------------------------------------------condição erro de muitos problemas
    if len(problems) > 5:
        return "Error: Too many problems."
    #------------------------------------------------------------------cria racional que analisa os items da lista
    else:
        Out = []
        for p in problems:
            s = p.split() 
    #------------------------------------------------------------------condição para o erro de numeros com outros caracteres        
            if re.search('[^0-9]', s[0]) or re.search('[^0-9]', s[2]):
                return "Error: Numbers must only contain digits."    
    #------------------------------------------------------------------condição para o erro de operador errado          
            elif re.search('[*/]', s[1]):
                return "Error: Operator must be '+' or '-'."       
    #------------------------------------------------------------------condição para o erro de numero grande      
            elif len(s[0]) > 4 or len(s[2]) > 4:
                return "Error: Numbers cannot be more than four digits."
            else:            
              big = str(max(int(s[0]),int(s[2])))
              r = ops[s[1]](int(s[0]),int(s[2]))
              Se = ((len(big)+ 2)*'-')
              t = len(Se)                         
              o = (f"{s[0] : >{t}}\n{s[1] : <1}{s[2]: >{t-1}}\n{Se : >{t}}\n{r : >{t}}")
              Out.append(o)
  
    return print((Out[0],Out[1]))
print(arithmetic_arranger(problems))
```
